# Ques 1.1:- Plot the bar graph of number of restaurants present in Delhi NCR vs Rest of India.

```python
#Delhi-NCR  vs rest of India

import matplotlib.pyplot as plt # python library form plotting graphs
import numpy as np  # python library for array-processing  
import pandas as pd # python library for data processing, CSV file I/O
from datetime import datetime
# read the zomato csv file using pandas
India=pd.read_csv('zomato.csv',encoding='latin-1')
df=India
# selecting only Indian restaurents Country code for India is 1
df=df[df['Country Code']==1]
# selecting the cities which comes under Delhi_NCR so that we can get the dataset of Delhi-NCR restro 
Delhi_NCR=df[(df.City=='New Delhi') | (df.City=='Gurgaon')  | (df.City=='Noida') | (df.City=='Faridabad') | (df.City=="Ghaziabad")]
# Dataset for rest of India
rest_of_India=df[(df.City!='New Delhi') & (df.City!='Gurgaon') & (df.City!='Noida') & (df.City!='Faridabad') & (df.City!="Ghaziabad")]
# lets find the no of restro  in Delhi-NCR
restro_in_Delhi_NCR=(Delhi_NCR['City'].count())
# now find the no of restro in rest of India
restro_in_rest_of_India=rest_of_India.City.count()

# plotting the bar graph for number of restaurants present in Delhi NCR vs Rest of India.

# data for x-axis of the graph
x=['restro_in_Delhi_NCR','restro_in_rest_of_India']
# data for y-axis of graph
y=[restro_in_Delhi_NCR,restro_in_rest_of_India]
# plotting the graph
plt.bar(x,y, color = 'b',edgecolor='black',width=0.5)
# providing the x-label
plt.xlabel('Location')
# y-label
plt.ylabel('No of restros')
# title of the graph
plt.title('Graph of Delhi_NCR v/s Rest of India')
# priting the value on the top of the bars using enumerate()...enumerate() method adds a counter to an iterable and returns it in a form of enumerate object. 
for index, value in enumerate(y):

    plt.text( index, value,str(value))
# show the graph
plt.show()
```

```python
# code for feature 2
import matplotlib.pyplot as plt # python library form plotting graphs
import numpy as np  # python library for array-processing  
import pandas as pd # python library for data processing, CSV file I/O
from datetime import datetime # lib for date time

# read the csv file
India=pd.read_csv('zomato.csv',encoding='latin-1')
df=India
# lets take the Indain retro from the data-set by using country code of India which is 1
df=df[df['Country Code']==1]
# lets sort the top 50 restro from the dataset on the bases of the aggeragate rating of the restro
top_50=(df['Aggregate rating']).sort_values(ascending=False)[0:50]
# storing the indexes of these 50 restro to make a new dataset of these restro so that i can move further to do analysis
top_50index=(top_50.index)
# required dataset
df=df.loc[top_50index]


# lets make a new dataframe having only Aggregate Rating of retro and no of Cuisines served by retro 
no_of_cuisines=[]
for i in df.Cuisines:
    a=i.split(',')
    # count the len for no of cuisines
    no_of_cuisines.append(len(a))
no_of_cuisines_series=pd.Series(no_of_cuisines,name='C_count')# mafe a series of count of no of cuisines served
Rating=(df['Aggregate rating']).reset_index(drop=True)# agg rating ki series

new_df = Rating.to_frame().join(no_of_cuisines_series)# new data frame having Aggregate rating and Count of cuisines served
# Graph1 : This graph shows the count of cuisines served vs rating of the restro(sabhi 50 restro ka)
plt.scatter(no_of_cuisines_series,Rating,color='green',s=Rating*5)
plt.xlabel('No of cuisines served')
plt.ylabel('Ratings')
plt.title('Graph 1:Aggregate rating  vs No of Cuisines served',fontweight="bold")
plt.show()

# Here I had made another graph in which i had tried to made a graph showing the average aggregate rating on the
# bases of count of cuisines served for eg if 3 cuisines are served by 10 restro (out of these top 50 restro) then what's their
# average agg. rating:
cuisines_total_rating={}# yeh dict mei maine no of cuisines served (1,2,3 or n) as keys li hai and unke respective rating ka sum as values
# means yeh dict bta rhi hai ki agr 10 restro 3 cuisines serve kar rhe hai toh unka total rating kitna hai
# (I need this to finally count their avg aggregate rating)
for i in range(len(no_of_cuisines)):
    # if no of cuisine is in the list then add the rating of the restro with that no of cusines else initiate it with the rating 
    if no_of_cuisines[i] in cuisines_total_rating:
        cuisines_total_rating[no_of_cuisines[i]]+=Rating[i]
    else:# initilizing the rating with a no of cuisine say is no of cusine is 1 then its inital rating is initiate here
        cuisines_total_rating[no_of_cuisines[i]]=Rating[i]
# Now I made a dict which have the keys as the no of cusines served and values as how many restro served x no of cuisines
cuisines_count={}# yeh dict bta rhi hai ki kitne retro (1,2,3,or n ) no of cuisines serve kar rhe hai 
for i in range(len(no_of_cuisines)):
    #if no of cuisine is in the list then inc the count of the restro by 1 else initiate it with 1
    if no_of_cuisines[i] in cuisines_count:
        cuisines_count[no_of_cuisines[i]]+=1
    else:
        cuisines_count[no_of_cuisines[i]]=1
# now the final dict having the average rating of the no of cuisnies served by x no of restro
final_dict={}# yeh dict bta rhi hai ki agr restro sirf 1 cusine serve karta hai toh uski avg rating kitni aa rhi hai agr 2 karta
# hai toh kitni and so on 
for i in cuisines_count.keys():
     
    final_dict[i]=cuisines_total_rating[i]/cuisines_count[i]
# made two list of no of cuisines served by retroes  and their count 
No_of_cuisines=list(final_dict.keys())
Average_rating=list(final_dict.values())
# lets plot the graph form the above analysis
plt.scatter(No_of_cuisines,Average_rating,color='green')
plt.xlabel('No of cuisines served')
plt.ylabel('Ratings')
plt.title('Graph 2:Average Aggregate rating  vs No of Cuisines served',fontweight="bold")
plt.show()
```

# QUES 1.5.5 :- Find the weighted restaurant rating of each locality and find out the top 10 localities with more weighted restaurant rating?

    Weighted Restaurant Rating= (number of votes * rating) /  (number of votes) .
                                                                                                                                                    
here in the ans of this problem I had printed weighted restaurant rating of each locality and made a bar graph of top 10 localities with more weighted restaurant rating   and then print them                                                                                                                                                  
                                                                                                                                          

```python
# code for 1.5.5
import matplotlib.pyplot as plt # python library form plotting graphs
import numpy as np  # python library for array-processing  
import pandas as pd # python library for data processing, CSV file I/O
from datetime import datetime # lib for date time

# read the csv file
India=pd.read_csv('zomato.csv',encoding='latin-1')
df=India
# lets take the Indain retro from the data-set by using country code of India which is 1
df=df[df['Country Code']==1]
# dropping the restro having votes and rating==0
df=df[(df.Votes!=0)]
df=df[df['Aggregate rating']!=0]
# reseting the index
df=df.reset_index(drop=True)
# lets find the LOcality weighted restaruent rating

# here the task is to find the weight rating of restro ie(Weighted Restaurant Rating= (number of votes * rating) /  (number of votes) .)
# for that I have to  first find the  (number of votes * rating) of the restro and this is stored in the Locality_numerator={} 
Locality_numerator={}
# now I had to cal the  (number of votes) for the restro and I stored that in Locality_deno{} dict
Locality_deno={}
# lets itrate over the localities and calculate the Locality_numerator and Locality_deno for every city in INDIA 
for i in range(len(df)):
    if  df.Locality[i] in Locality_numerator:
        # if lcoality is in the dict then inc the value by votes*rating otherwise initate the value
        Locality_numerator[df.Locality[i]]+=(df.Votes[i]*df['Aggregate rating'][i])
    else:
        Locality_numerator[df.Locality[i]]=(df.Votes[i]*df['Aggregate rating'][i])
for i in range(len(df)):
    # if locality is in the dict then inc the value by votes otherwise initate the value
    if  df.Locality[i] in Locality_deno:
        Locality_deno[df.Locality[i]]+=df.Votes[i]
    else:
        Locality_deno[df.Locality[i]]=df.Votes[i]
# now for WEIGHTED resto rating divide the numerator by deno of each lcoality resp and store the result in final dict
Final_dict={}
for i in Locality_numerator.keys():
    Final_dict[i]=Locality_numerator[i]/Locality_deno[i]
# ANS for  **Find the  weighted restaurant rating of each locality ****
print(Final_dict)


# for top 10 locality with more weighted restaurant rating
# lets sort the Final_dict and made a list of tuples having locality and their weighted restro rating
b = sorted(Final_dict.items(), key=operator.itemgetter(1),reverse=True)
# made a list of top 10 locality and weighted restro rating
Top_10_Localities=[]
top_10_Localities_rating=[]
for i in range(15):
    Top_10_Localities.append(b[i][0])
    top_10_Localities_rating.append(b[i][1])
    
# plotting the graph for top 10 locality with more weighted restaurant rating
# plot a bar graph
plt.bar(Top_10_Localities,top_10_Localities_rating, color = 'b', width = 0.5,edgecolor='black')
# privided the label and font size
plt.xlabel('Top 10 localities',fontsize=30,fontweight='bold')
plt.ylabel('Weighted Rating',fontsize=30,fontweight='bold')
plt.xticks(rotation=90)
# priting the value on the top of the bars using enumerate()...enumerate() method adds a counter to an iterable and returns it in a form of enumerate object. 
for index, value in enumerate(top_10_Localities_rating):

    plt.text( index, value,str(value))
plt.grid()
# title+
plt.title('graph of top 10 Localities have a maximum weighted Rating',fontweight="bold")
plt.tick_params(labelsize=20)
plt.show()

# ANS for **top 10 locality with more weighted restaurant rating***
print(Top_10_Localities)
```

```python
#Plot the bar graph top 15 restaurants have a maximum number of outlets(Consider only Indian restaurants in this analysis -as mention in heading)
import matplotlib.pyplot as plt # python library form plotting graphs
import numpy as np  # python library for array-processing  
import pandas as pd # python library for data processing, CSV file I/O
import operator
from datetime import datetime
# read the zomato csv file using pandas
India=pd.read_csv('zomato.csv',encoding='latin-1')
df=India
# selecting only Indian restaurents Country code for India is 1
df=df[df['Country Code']==1]
# Here a dict is made in which the keys is name of OUTLETS and value is THEIR  NO OF COUNT
No_of_outlets={}
# for this dict we have to itrate over the Restro names and check their counts
for i in df['Restaurant Name']:
    # if name in dict then inc the count otherwise initilize it
    if i in No_of_outlets:
        No_of_outlets[i]+=1
    else:
        No_of_outlets[i]=1
# lets sort the dict and made a list of tuples having the outlets and their resp counts
b = sorted(No_of_outlets.items(), key=operator.itemgetter(1),reverse=True)
# now made a list of top 15 restor having the oulets and their resp count using the list of tuples ie 'b'
Top_15_restro=[]
No_of_oulet_of_top_15_restro=[]
# list of top 15 restro having max outlets
for i in range(15):
    Top_15_restro.append(b[i][0])
    No_of_oulet_of_top_15_restro.append(b[i][1])
# Plotting the bar graph top 15 restaurants have a maximum number of outlets
plt.bar(Top_15_restro,No_of_oulet_of_top_15_restro, color = 'b', width = 0.5,edgecolor='black')
# providing the xlabel and ylabel
plt.xlabel('Name of restro',fontweight='bold',fontsize=30)
plt.ylabel('No of restros outlets',fontweight='bold',fontsize=30)
plt.xticks(rotation=90)
# priting the value on the top of the bars using enumerate()...enumerate() method adds a counter to an iterable and returns it in a form of enumerate object. 
for index, value in enumerate(No_of_oulet_of_top_15_restro):

    plt.text( index, value,str(value),fontsize=20)
plt.grid()
# providing the size of labels
plt.tick_params(labelsize=30)
plt.title('graph of top 15 restaurants have a maximum number of outlets',fontweight="bold",fontsize=30)
plt.show()
```

# 5) Plot the bubble graph of a number of Restaurants present in the city of India and keeping the weighted restaurant rating of the city in a bubble.

```python
#Plot the bubble graph of a number of Restaurants present in the city of India and keeping the weighted restaurant rating of the city in a bubbl

import matplotlib.pyplot as plt # python library form plotting graphs
import numpy as np  # python library for array-processing  
import pandas as pd # python library for data processi ng, CSV file I/O
import operator
from datetime import datetime
# read the zomato csv file using pandas
India=pd.read_csv('zomato.csv',encoding='latin-1')
df=India
# selecting only Indian restaurents Country code for India is 1
df=df[df['Country Code']==1]
# droping the restro having the rating and votes 0
df=df[(df.Votes!=0)]
df=df[df['Aggregate rating']!=0]
# reseting the indexes
df=df.reset_index(drop=True)
# here the task is to find the weight rating of restro ie(Weighted Restaurant Rating= (number of votes * rating) /  (number of votes) .)
# for that I have to  first find the  (number of votes * rating) of the restro and this is stored in the City_numerator={} dict
City_numerator={}
# now I had to cal the  (number of votes) for the restro and I stored that in City_deno{} dict
City_deno={}
# lets itrate over the cities and calculate the city_numerator and city_deno for every city in INDIA 

for i in range(len(df)):
    # if city is in the dict then inc the value by votes*rating otherwise initate the value
    if  df.City[i] in City_numerator:
        City_numerator[df.City[i]]+=(df.Votes[i]*df['Aggregate rating'][i])
    else:
        City_numerator[df.City[i]]=(df.Votes[i]*df['Aggregate rating'][i])
for i in range(len(df)):
    # if city is in the dict then inc the value by the no of votes otherwise initiate it
    if  df.City[i] in City_deno:
        City_deno[df.City[i]]+=df.Votes[i]
    else:
        City_deno[df.City[i]]=df.Votes[i]
# now for WEIGHTED resto rating divide the numerator by deno of each city resp and store the result in final dict
Final_dict={}
for i in City_numerator.keys():
    Final_dict[i]=City_numerator[i]/City_deno[i]
# lets make a list having the vales of final dict in it ie weighted restro ratings
Weight_restro_rating=[]
for i in Final_dict.values():
    Weight_restro_rating.append(i*100) #here the rating is mult by 100 beacuse of :-the bubbles should be clearly visible 
# lets find the City and no of restro present in that city
# for that I will itrate over the dataset having INDIAN restro and then count the no of restro in the specific city
# and save the result in the City_restro_count
City_restro_count={}
for i in df.City:
    
    if i in City_restro_count:
        City_restro_count[i]+=1
    else:
        City_restro_count[i]=1
# made list of cities and no of restero in the cities
Cities=[]
Count=[]
for i in (City_restro_count.keys()):
    Cities.append(i)
    Count.append(City_restro_count[i])

# plot a bubble graph having cities and no of restro and their weighted restro rating as the size of bubble(*100)
plt.scatter(Cities,Count, color = 'red',s=Weight_restro_rating)# buble graph
plt.xlabel('Cities',fontweight='bold',fontsize=30)# provided the xlabel and the font size
plt.ylabel('Count',fontweight='bold',fontsize=30)# provided the ylabel and the font size
plt.xticks(rotation=90)
plt.grid()
plt.tick_params(labelsize=30)# to fix the size of labels
plt.rcParams["figure.figsize"] = (30,20)
plt.title('graph of Cities vs Count with bubble as W.R.R',fontweight="bold",fontsize=40)# # provided the title and the font size
plt.show()
```
